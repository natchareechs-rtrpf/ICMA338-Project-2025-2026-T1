# Finding Error in Data by using Gradient Descent and Linear Regression

Purpose is to see error of our data by using those 2 methods.

We want to define Cost Function to minimize our function. So, we are going to do that with our data set we got to see the error.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
data = np.array([
    [4, 6, 0],
    [2, 6, 2],
    [6, 2, 2],
    [3, 4, 3]
])
x1 = data[:, 0]
x2 = data[:, 1]
y  = data[:, 2]

# MSE (Mean Square Error)

def loss_function(params, x1, x2, y):
    w1, w2, c = params
    y_pred = w1*x1 + w2*x2 + c
    return np.mean((y - y_pred) ** 2)

def gradient(params, x1, x2, y):
    w1, w2, c = params
    y_pred = w1*x1 + w2*x2 + c
    error = y - y_pred

    dw1 = -2 * np.mean(x1 * error)
    dw2 = -2 * np.mean(x2 * error)
    dc  = -2 * np.mean(error)
    return np.array([dw1, dw2, dc])

params = np.array([0.0, 0.0, 0.0])
learning_rate = 0.01 #Prevent it to overshooting
n_iterations = 5000  #Repeat it 5000 times

for i in range(n_iterations):
    grads = gradient(params, x1, x2, y)
    params -= learning_rate * grads

w1_opt, w2_opt, c_opt = params
print(f"Optimized parameters: w1 = {w1_opt:.2f}, w2 = {w2_opt:.2f}, c = {c_opt:.2f}")

y_fit = w1_opt * x1 + w2_opt * x2 + c_opt
sum_x = x1 + x2

plt.plot(sum_x, y, '.r', markersize=15, label='Data')
plt.plot(sum_x, y_fit, 'b.', markersize=10, label='Prediction')
for i in range(len(y)):
    plt.plot([sum_x[i], sum_x[i]], [y[i], y_fit[i]], 'k-', alpha=0.3)

plt.xlabel('Sum of Inputs')
plt.ylabel('Y')
plt.legend()
plt.title('Linear Regression using Gradient Descent')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
data = np.array([
    [4, 6, 0],
    [2, 6, 2],
    [6, 2, 2],
    [3, 4, 3]
])

dx = data[:, 0] + data[:, 1]
dy = data[:, 2]

m = 0
c = 0
alpha = 0.01          # Learning Rate
iterations = 10000    # Steps
n_points = float(len(dx))
history = []

# Gradient Descent
for i in range(iterations):
    y_pred = m * dx + c

    # Calculate gradients
    D_m = (-2/n_points) * sum(dx * (dy - y_pred))
    D_c = (-2/n_points) * sum(dy - y_pred)

    # Update parameters
    m = m - alpha * D_m
    c = c - alpha * D_c

    # Save data for every 100 steps
    if i % 100 == 0:
        history.append((m,c))

# Plotting
plt.figure(figsize=(8, 6))
plt.plot(dx, dy, '.r', markersize=15, label='Data points')

x_line = np.linspace(min(dx)-1, max(dx)+1, 100)

# Plot history (Green lines)
for i, (m_h, c_h) in enumerate(history):
    y_line = m_h * x_line + c_h
    plt.plot(x_line, y_line, 'g-', alpha=0.1 + 0.5*(i/len(history)))

# Plot final line
y_line = m * x_line + c
plt.plot(x_line, y_line, 'b-', linewidth=3, label=f'Final line (Iter {iterations})')

plt.title(f"Gradient Descent\nFinal line: y = {m:.2f}x + {c:.2f}")
plt.xlabel('Sum of Inputs')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

print(f"Final m: {m}")
print(f"Final c: {c}")
```
